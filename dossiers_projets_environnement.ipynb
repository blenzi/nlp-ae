# Récupération de dossiers et avis de l'Ae depuis projets-environnement.fr

```python
import os
import urllib.request
from pathlib import Path
import pandas as pd
import numpy as np
import requests
```

```python
def file_size(url):
    "Retrieve file size in MB from URL without downloading it"
    try:
        req = requests.head(url)
        # if req and 'content-length' in req.headers:  
        return int(req.headers["content-length"])/1e6
    except:
        return 0

def download(url, dest_path='.', dest_name=None):
    "Download file to destination and return True if succeeded" 
    if dest_name is None:
        dest_name = Path(url).name.replace('%20', ' ')
    dest = Path(dest_path)/dest_name
    if dest.exists():
        return True
    try:
        urllib.request.urlretrieve(url.replace(' ', '%20'), dest)
        return True
    except:
        return False
```

### Liste des entrées avec lien vers l'Ae non nul

```python
url = 'https://www.projets-environnement.gouv.fr/api/v2/catalog/datasets/projets-environnement-diffusion/exports/csv?where=dc_relation_expertise_avisae%20IS%20NOT%20NULL&limit=-1&offset=0&timezone=UTC'
df = pd.read_csv(url, sep=';')
with pd.option_context('display.max_columns', None):
    display(df)
```

### En Bretagne

```python
bretagne = df['dc_location'].str.contains('22|29|35|56')
with pd.option_context('display.max_columns', None):
    display(df[bretagne])
```

### Taille des fichiers en MB: avis, dossier, étude d'impact

```python
columns = ['dc_relation_expertise_avisae', 'dc_relation_officialdocument', 'dc_relation_expertise_etudeimpact']
fs = df.loc[bretagne, columns].head(10).applymap(file_size)
with pd.option_context("display.precision", 3):
    display(fs)
```

```python
with pd.option_context('display.max_colwidth', None):
    display(df.loc[fs[fs['dc_relation_expertise_avisae'] < 0.1].index, 'dc_relation_expertise_avisae'])
```
